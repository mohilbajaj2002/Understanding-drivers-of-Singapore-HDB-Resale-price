# Extracting MRT Geolocation data

```python
import xml.etree.ElementTree as ET
filePath = r'G_MP08_ACT_MRT_STN_PT.kml'
tree = ET.parse(filePath) 
lineStrings = tree.findall('Point')

print(lineStrings)
for attributes in lineStrings:
    for subAttribute in attributes:
        if subAttribute.tag == 'coordinates':
            print (subAttribute.tag, subAttribute.text)
```

```python
from bs4 import BeautifulSoup as bs

with open('G_MP08_ACT_MRT_STN_PT.kml') as data:
    kml_soup = bs(data, 'lxml-xml') # Parse as XML

lat = []
long = []

descriptions = kml_soup.find_all('Point')
for d in descriptions:
    coord = d.find_all('coordinates')
    for c in coord:
        values = c.text.split(',')
        latitude = values[1]
        longitude = values[0]
        lat.append(latitude)
        long.append(longitude)
   
print (len(lat),len(long))
```

# Extracting hawker centre geolocation data

```python
from bs4 import BeautifulSoup as bs

with open('hawker-centres-kml.kml') as data:
    kml_soup = bs(data, 'lxml-xml') # Parse as XML
    
add = []
lat = []
long = []

descriptions = kml_soup.find_all('ExtendedData')
for d in descriptions:
    coord = d.find_all('SimpleData')
    address = coord[0].text
    latitude = coord[3].text
    longitude = coord[8].text
    print(address, latitude,longitude)
    add.append(address)
    lat.append(latitude)
    long.append(longitude)
   
print (len(add),len(lat),len(long))
```

```python
hawker = pd.DataFrame()
hawker['address'] = add
hawker['latitude'] = lat
hawker['longitude'] = long

hawker.head()
```
